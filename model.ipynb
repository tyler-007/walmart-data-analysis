```python
city_avg = data.groupby('City').agg({'Rating': 'mean', 'Quantity': 'mean', 'Unit price': 'mean'}).reset_index()

# Plot average rating, quantity, and unit price grouped by city
fig, axes = plt.subplots(3, 1, figsize=(10, 15))

# Plot average rating
axes[0].bar(city_avg['City'], city_avg['Rating'], color='red')
axes[0].set_title('Average Rating by City')
axes[0].set_ylabel('Rating')

# Plot average quantity
axes[1].bar(city_avg['City'], city_avg['Quantity'], color='green')
axes[1].set_title('Average Quantity by City')
axes[1].set_ylabel('Quantity')

# Plot average unit price
axes[2].bar(city_avg['City'], city_avg['Unit price'], color='blue')
axes[2].set_title('Average Unit Price by City')
axes[2].set_ylabel('Unit Price')

# Adjust layout
plt.tight_layout()
plt.show()
```

```python
city_product_quantity = data.groupby(['City', 'Product line'])['Quantity'].sum().reset_index()

# Get unique cities in the dataset
cities = city_product_quantity['City'].unique()

# Iterate over each city and plot the quantity of different products
for city in cities:
    # Filter data for the current city
    city_data = city_product_quantity[city_product_quantity['City'] == city]
    
    # Plot
    plt.figure(figsize=(10, 6))
    plt.bar(city_data['Product line'], city_data['Quantity'], color='skyblue')
    plt.title(f'Quantity of Products in {city}')
    plt.xlabel('Product Line')
    plt.ylabel('Quantity')
    plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for better visibility
    plt.tight_layout()
    plt.show()
```

```python
data['Time'] = pd.to_datetime(data['Time'])

# Extract hour from 'Time'
data['Hour'] = data['Time'].dt.hour

# Group by hour and product line, count occurrences, and find the most bought product line for each hour
most_bought_per_hour = data.groupby(['Hour', 'Product line']).size().reset_index(name='Count')
most_bought_per_hour = most_bought_per_hour.loc[most_bought_per_hour.groupby('Hour')['Count'].idxmax()]

# Plotting
plt.figure(figsize=(12, 6))
bars = plt.bar(most_bought_per_hour['Hour'], most_bought_per_hour['Count'], color='skyblue')
plt.title('Most Bought Product Line for Each Hour of the Day')
plt.xlabel('Hour of the Day')
plt.ylabel('Count')
plt.xticks(range(24))
plt.grid(axis='y', linestyle='--', alpha=0.7)

for bar, product_line in zip(bars, most_bought_per_hour['Product line']):
    plt.text(bar.get_x() + bar.get_width() / 2, bar.get_height(), product_line, 
             ha='center', va='bottom', rotation=90, fontsize=8)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Assuming your DataFrame is named data

# Convert 'Date' column to datetime format
data['Date'] = pd.to_datetime(data['Date'])

# Extract month and year from 'Date'
data['Month'] = data['Date'].dt.month
data['Year'] = data['Date'].dt.year

# Group the data by city, product line, month, and year, and calculate total sales
city_product_monthly_sales = data.groupby(['City', 'Product line', 'Year', 'Month'])['Total Sales'].sum().reset_index()

# Plot line charts for city-wise sales trends for different products
for city in city_product_monthly_sales['City'].unique():
    city_data = city_product_monthly_sales[city_product_monthly_sales['City'] == city]
    for product_line in city_data['Product line'].unique():
        product_data = city_data[city_data['Product line'] == product_line]
        
        plt.figure(figsize=(8, 6))
        sns.lineplot(data=product_data, x='Month', y='Total Sales', style='Year', markers=True)
        plt.title(f'Sales Trends for {product_line} in {city}')
        plt.xlabel('Month')
        plt.ylabel('Total Sales')
        plt.xticks(range(1, 13))  # Assuming data spans all 12 months
        plt.legend(title='Year', bbox_to_anchor=(1.05, 1), loc='upper left')
        plt.grid(True)
        plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Assuming your DataFrame is named data

# Convert 'Date' column to datetime format
data['Date'] = pd.to_datetime(data['Date'])

# Extract month and year from 'Date'
data['Month'] = data['Date'].dt.month
data['Year'] = data['Date'].dt.year

# Group the data by city, product line, month, and year, and calculate total sales
city_product_monthly_sales = data.groupby(['City', 'Product line', 'Year', 'Month'])['Total Sales'].sum().reset_index()

# Plot line charts for city-wise sales trends for all product lines together
for city in city_product_monthly_sales['City'].unique():
    city_data = city_product_monthly_sales[city_product_monthly_sales['City'] == city]
        
    plt.figure(figsize=(10, 6))
    sns.lineplot(data=city_data, x='Month', y='Total Sales', hue='Product line', style='Product line', markers=True)
    plt.title(f'Sales Trends in {city} for All Product Lines')
    plt.xlabel('Month')
    plt.ylabel('Total Sales')
    plt.xticks(range(1, 13))  # Assuming data spans all 12 months
    plt.legend(title='Product Line', bbox_to_anchor=(1.05, 1), loc='upper left')
    plt.grid(True)
    plt.show()
```
